We are going to use:
> pyautogui.click -> Click
> pyautogui.press -> Press any key
> pyautogui.hotkey -> Key's sets
> py.autogui.write -> Write text

```python
import time as t
import pyautogui as py
from selenium import webdriver
import pyperclip 

#Function to add a 1s delay between lines.
py.PAUSE = 0.3

# Step 0 - Alet
py.alert("Attention! Don't use the computer at this moment")

# Step 1 - Acess the company's system
# Initialize Chrome Browser
browser = webdriver.Chrome()

# Open a specific website
browser.get("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")

# Maximize window
browser.maximize_window()

# Step 2 - Login
# Login
py.click()
py.press('tab')
py.typewrite('Mateus of Saints Gentleman', interval=0.1)

# Password
py.press('tab')
py.typewrite('LittleGentleman123', interval=0.1)

# Press Enter key
py.press('tab')
py.press('enter')

# Step 3 - Download the data base
t.sleep(2)
py.click(x=284, y=339, clicks=2)
py.click(x=1107, y=172)
```
