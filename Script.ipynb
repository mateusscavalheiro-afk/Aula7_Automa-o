We are going to use:

> pyautogui.hotkey -> Key's sets
> py.autogui.write -> Write text

```python
import time as t
import pyautogui as py
from selenium import webdriver
import pyperclip 

#Function to add a 1s delay between lines.
py.PAUSE = 0.5

# Step 0 - Alet
py.alert("Attention! Don't use the computer at this moment")

# Step 1 - Access the company's system
# Initialize Chrome Browser
browser = webdriver.Chrome()

# Open a specific website
browser.get("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")

# Maximize window
browser.maximize_window()

# Step 2 - Login
# Login
py.click()
py.press('tab')
py.typewrite('Mateus of Saints Gentleman', interval=0.1)

# Password
py.press('tab')
py.typewrite('LittleGentleman123', interval=0.1)

# Press Enter key
py.press('tab')
py.press('enter')

# Step 3 - Download the data base
t.sleep(2)
py.click(x=284, y=339, clicks=2)
t.sleep(3)
py.click(x=1107, y=172)
t.sleep(1)
py.alert('Archive downloaded? Press OK to confirm')
print('Archive successfully downloaded!')
t.sleep(3)
py.hotkey('alt', 'f4')
```

```python
'''
%pip install pandas
%pip install openpyxl
%pip install numpy'''

# Step 4 - Calculate indicators
# Import data base
import pandas as pd

table = pd.read_csv(r"C:\Users\mateus_s_cavalheiro\Downloads\Compras.csv", sep=";")

display(table)

'''
Total Spent -> Sum of all collumn (ValorFInal)
Qtd. Total Produces -> Sum of all collumn (Quantidade)
Averege price -> ValorFinal / Quantidade
'''

total_spent = table["ValorFinal"].sum()
quantity = table["Quantidade"].sum()
averege_price = total_spent / quantity

print("Total Spent -> R$",total_spent)
print("Total Items -> ",quantity, "items")
print("Averege Price -> R$", averege_price, "per item")

print("\nData successfully collected!")
py.alert("Data successfully collected!")    
```

```python
'''%pip install python-dotenv'''

# Import dotenv to load .env configurations
import os
import time as t
import pyautogui as py
from selenium import webdriver
from dotenv import load_dotenv as ldv
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

py.PAUSE = 1

# Load config. from .env
ldv()
email = os.getenv("email")
password = os.getenv("password")

chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_experimental_option("detach", True)
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
servicey = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=servicey, options=chrome_options)
browser.maximize_window()

# Step 5 - Send treated e-mail to other section or diretory

# Enter in new e-mail
browser.get("https://mail.google.com/mail")
py.write(email)
py.press('enter')
t.sleep(7)
py.write(password)
py.press('enter')
t.sleep(8)

# Click "Write"
py.press('tab', presses=14, interval=0.1)
py.press('enter')
t.sleep(8)

# Fill "To"
py.write("email_to_recipient") 
py.press('enter')
py.press('tab')
t.sleep(1)

# Fill "Subject"
py.write("Subject Text")
py.press('tab')
t.sleep(1)

# Fill "E-mail corpse"
py.write("E-mail corpse text")
t.sleep(1)

# Send e-mail
py.press('tab')
t.sleep(1)
py.press('enter')
```
